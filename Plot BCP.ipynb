```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
likelihood = [2,3,3.5,4]
Impact = [4.5,3,4,3.5]
```

```python
# Create an axes a figure object
f, ax = plt.subplots(1, figsize=(5, 5), dpi=170)


# Scatter the points 
#ax.scatter([3.5, 3], [3, 4], s=80)

ax.scatter(x = Impact, y = likelihood , s=50)

# axis limits
plt.xlim(1, 5)
plt.ylim(1,5)

#labels
ax.set_ylabel("Likelihood")
ax.set_xlabel('Extend of Impact')

# activate grid
ax.grid("on")

# Add three text patches to the plot
ax.text(2.6, 3.1, 'Loss of access',  fontsize=8)
ax.text(3.1, 4.1, 'Loss of technology',  fontsize=8)
ax.text(3.3, 3.6, 'Loss of building functions',  fontsize=8)
ax.text(4, 2.1, 'Loss of building',  fontsize=8)

# save as jpeg
plt.savefig('DynamisBCP.jpeg',transparent=True, dpi=300, bbox_inches='tight')
```
